```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

```python

# Define the model
class SineNet(nn.Module):
    def __init__(self):
        super(SineNet, self).__init__()
        self.fc1 = nn.Linear(1, 50)  # Input layer
        self.fc2 = nn.Linear(50, 50) # Hidden layer
        self.fc3 = nn.Linear(50, 1)  # Output layer

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Generate data
x = torch.unsqueeze(torch.linspace(-5*np.pi, 5*np.pi, 2000), dim=1)
y = torch.sin(x)

# Model, optimizer and loss function
model = SineNet()
optimizer = optim.Adam(model.parameters(), lr=0.01)
criterion = nn.MSELoss()
```

```python

# Training loop
for epoch in range(2000):
    prediction = model(x)
    loss = criterion(prediction, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 200 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item()}')
```

```python

# Test the model
with torch.no_grad():
    test_x = torch.unsqueeze(torch.linspace(-2*np.pi, 2*np.pi, 400), dim=1)
    test_y = model(test_x)
    import matplotlib.pyplot as plt
    plt.plot(test_x.numpy(), test_y.numpy(), label='Model Prediction')
    plt.plot(test_x.numpy(), np.sin(test_x.numpy()), label='True Function')
    plt.legend()
    plt.show()
```
