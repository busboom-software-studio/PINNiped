```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-15*np.pi, 15*np.pi, 1000)
y = np.sin(x + np.sin(x/4))

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x + sin(x/4))')
plt.title('Sin Wave with Variable Period')
plt.show()
```
